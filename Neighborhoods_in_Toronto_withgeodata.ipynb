```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
table = soup.find('table',{'class':'wikitable sortable'})
table
```

```python
PostalCode = []
Borough = []
Neighborhood = []
for row in table.findAll("tr")[1:]:
    data = row.findAll("td")
    PostalCode.append(data[0].find(text=True))
    Borough.append(data[1].find(text=True))
    Neighborhood.append(data[2].find(text=True))
print(PostalCode)
print(Borough)
print(Neighborhood)
```

```python
import pandas as pd
nh_df = pd.DataFrame()
nh_df['PostalCode'] = PostalCode
nh_df['Borough'] = Borough
nh_df['Neighborhood'] = Neighborhood
nh_df
```

```python
nh_df = nh_df[nh_df.Borough != "Not assigned"]
nh_df
```

```python
grouped = nh_df.groupby(['PostalCode'])
for name, group in grouped:
    print(name)
    print(group)
  
```

```python
nh_df['Neighborhood'] = nh_df[['PostalCode','Borough','Neighborhood']].groupby(['PostalCode'])['Neighborhood'].transform(lambda x: ','.join(x))
nh_df[['PostalCode', 'Borough', 'Neighborhood']].drop_duplicates()
```

```python
import numpy as np
nh_df.replace(r'Not assigned', np.nan, regex=True, inplace=True)
nh_df
```

```python
df = pd.read_csv("http://cocl.us/Geospatial_data", index_col='Postal Code')
df
```

```python
for index, row in nh_df.iterrows():
    postalcode = row['PostalCode']
 
    latitude = df.at[postalcode, 'Latitude']
    longitude = df.at[postalcode, 'Longitude']
    nh_df.loc[index, 'Latitude'] = latitude
    nh_df.loc[index, 'Longitude'] = longitude
nh_df
```
